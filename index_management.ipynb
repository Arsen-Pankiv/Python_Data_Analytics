```python
"let's now find the row with area name such as Brent"

rf_brent = rf[rf.area == "brent"]

rf_brent
 
 
```

```python
"""but we don't have index in order; let now fix it, it doesn't go one by one"""

rf_brent.reset_index(inplace=True)

"""now we have the value index back in place"""
rf_brent

```

```python
"""the folling method will create an index for the column"""

rf.index.name = "value_index"
rf_brent = rf[rf.area == "brent"]
rf_brent.reset_index(inplace=True)
rf_brent.set_index("value_index", inplace=True)

rf_brent
```

```python
"sorting"

title_salary_ctr = rf.pivot_table(
    values = "mean_salary",
    index="job_title",
    aggfunc=["mean","max","min"])

title_salary_ctr
```

```python
title_salary_ctr.sort_values(by=("mean", "mean_salary"), ascending= False)
```

```python
"we also may sort index in alphabetical order"

title_salary_ctr.sort_index(inplace=True)
title_salary_ctr
```

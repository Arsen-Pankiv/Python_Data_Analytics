Cleaning the data, let's filed Nan values with median or mean; let's also remove dupplicate values


```python
import pandas as pd
import matplotlib.pyplot as plt

rf = pd.read_csv("/Users/arsenpankiv/Desktop/Python_practice/small_dataset.csv")
rf["date"] = pd.to_datetime(rf["date"])
```

```python
rf.loc[0:15, "recycling_pct": "num_of_jobs"]
```

```python

"""let's now calculate median and impute it """
recycle1 = rf["recycling_pct"].median()
population1 = rf["population_size"].median()
job1 = rf["num_of_jobs"].median()
```

```python
"""let's now created new dateframe"""

df_filled = rf

"""lets use fillna for imoutation """

df_filled["recycling_pct"].fillna(recycle1)
df_filled["population_size"].fillna(population1)
df_filled["num_of_jobs"].fillna(job1)
```

```python
"""let's now save these changes"""
df_filled["recycling_pct"] = df_filled["recycling_pct"].fillna(recycle1)
df_filled["population_size"] = df_filled["population_size"].fillna(population1)
df_filled["num_of_jobs"] = df_filled["num_of_jobs"].fillna(job1)
```

```python
"""check"""

df_filled.loc[0:20, "recycling_pct":"num_of_jobs"]
```

```python
"""let's now remove duplciates of these three comlumns - duplicate rows removed using drop_duplicates;  let's also create a new dataframe"""

df_dup = df_filled

df_dup.drop_duplicates()

"""let's now save changes"""
df_dup = df_dup.drop_duplicates()



```

```python
"""let's now compare how many were removed"""

print(f"{len(df_dup)} - lenght of dupliocates removed")
print(f"{len(df_filled)} - lenght of df_filled dataframe removed")

```

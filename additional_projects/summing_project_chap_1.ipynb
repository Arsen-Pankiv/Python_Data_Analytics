```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast

# Load data
rf = pd.read_csv("/Users/arsenpankiv/Desktop/Python_practice/small_dataset.csv")

# Clean data
rf["date"] = pd.to_datetime(rf["date"])
rf["skills_list"] = rf["skills_list"].apply(
    lambda skills: ast.literal_eval(skills) if isinstance(skills, str) else skills)

#  taking one one role out of three

stat = rf[rf["job_title"] =="Statistician"]
```

```python
area_count = stat["area"].value_counts().head(8).to_frame()

import seaborn as sns

# we may also set the theme for our barchart 
sns.set_theme(style="whitegrid")

sns.barplot(data=area_count, y="area",x="count", hue="count", palette="light:purple", legend=False)
sns.despine()
plt.title("Number of areas per title")
plt.xlabel("Count of areas")
plt.ylabel("")
plt.show()
# to_frame transfroms it into a dataframe 
```

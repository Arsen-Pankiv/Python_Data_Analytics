```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import ast

# Load data
rf = pd.read_csv("/Users/arsenpankiv/Desktop/Python_practice/small_dataset.csv")

# Clean data
rf["date"] = pd.to_datetime(rf["date"])
rf["skills_list"] = rf["skills_list"].apply(
    lambda skills: ast.literal_eval(skills) if isinstance(skills, str) else skills)

stat = rf[rf["job_title"]== "Statistician"].copy()
```

```python

stat["mean_salary"].plot(kind="hist", bins=10, edgecolor= "black")



plt.xlim(0,45000)
"""let's npw format the x axe the way we did it previously"""
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"£{int(x/1000)}K"))
plt.title("Distibution of mean salary for Statistician")
plt.xlabel("Mean salary")
plt.ylabel("")
```

```python
import seaborn as sns 

sns.displot(stat["mean_salary"], kind="kde", fill=True)

# kind= kde makes smooth transition
# fill= True - fill out the background with the colour

plt.xlim(0,45000)
"""let's npw format the x axe the way we did it previously"""
ax = plt.gca()
ax.xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f"£{int(x/1000)}K"))
plt.title("Distibution of mean salary for Statistician")
plt.xlabel("Mean salary")
plt.ylabel("")
```
